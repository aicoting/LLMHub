```python
import torch
import torch.nn as nn
import math
import copy
```

## 1. Positional Encoding
+ ****`x``(batch_size, seq_len, d_model)`
+ ****
    - `PE`` (1,max_len,d_model)`  
    
    trick  
    ![](https://cdn.nlark.com/yuque/0/2025/png/28454971/1750153358121-7ea28063-685e-4fb8-bbfe-1b0a455ef5ba.png)
    - `PE` 
    - 
+ ****

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=5000):
        """
        

        
        - d_model: Transformertokenembedding
        - max_len: 5000
        """
        super().__init__()
        # 0 (max_len, d_model)
        pe = torch.zeros(max_len, d_model)  # 
        # 0 max_len-1 (max_len, 1)
        position = torch.arange(0, max_len, dtype=torch.float32).unsqueeze(1)

        # 
        # div_term (d_model // 2,) 2i,trick
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))

        # pe
        # 0,2,4,...sin(position / (10000^(2i/d_model)))
        pe[:, 0::2] = torch.sin(position * div_term)

        # pe
        # 1,3,5,...cos(position / (10000^(2i/d_model)))
        pe[:, 1::2] = torch.cos(position * div_term)

        #  batch  (1, max_len, d_model)
        #  shape  (batch_size, seq_len, d_model)
        pe = pe.unsqueeze(0)

        # buffer
        # 
        self.register_buffer('pe', pe)

    def forward(self, x):
        """
        

        
        - x:  (batch_size, seq_len, d_model)

        
        -  (batch_size, seq_len, d_model)
        """
        # 
        # self.pe[:, :x.size(1), :]  shape  (1, seq_len, d_model)
        x = x + self.pe[:, :x.size(1), :]
        return x
        
```

## 2.
+ ****`Q``K``V``(batch_size, head_num, seq_len,d_k)` `mask` 
+ ****
    - `scores = QKT/sqrt(d_k)``(batch, head, seq, seq)`
    - `-<font style="color:rgba(0, 0, 0, 0.8);"></font>` 
    -  softmax
    - `V`
+ ****
    - `(batch, head, seq, dk)`
    - 



```python
def scaled_dot_product_attention(q, k, v, mask=None):
    """
    
    - q:  (batch_size, num_heads, seq_len, d_k)
    - k:      (batch_size, num_heads, seq_len, d_k)
    - v:      (batch_size, num_heads, seq_len, d_k)
    - mask:  padding 

    
    - output:  (batch_size, num_heads, seq_len, d_k)
    - attn:  (batch_size, num_heads, seq_len, seq_len)
    """

    #  q/k/v  d_k
    d_k = q.size(-1)

    # Step 1: 
    # Q * K^T (batch_size, num_heads, seq_len, seq_len)
    #  i  j 
    scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(d_k)

    # Step 2: 
    if mask is not None:
        # 0 -infsoftmax0
        scores = scores.masked_fill(mask == 0, float('-inf'))

    # Step 3:  softmax
    #  token  key 
    attn = torch.softmax(scores, dim=-1)

    # Step 4:  V
    # V
    output = torch.matmul(attn, v)

    return output, attn
```

## 3. Multi-Head Attention
+ ****`Q,K,V``(batch,seq,d_model)` 
+ ****
    - `Q,K,V` ` d_k = d_model/num_heads`
    -  reshape  transpose`(batch, head_num, seq,d_k)` 
    - 
    - concatenate `(batch, seq, d_model)`
    - 
+ ****`(batch, seq, d_model)`


```python
class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads):
        """
        
        - d_model: 512
        - num_heads: 8

        d_model  num_heads
        """
        super().__init__()

        assert d_model % num_heads == 0  # 

        self.d_k = d_model // num_heads  #  key/query/value 
        self.num_heads = num_heads       # 

        #  QKV  d_model -> d_model
        self.w_q = nn.Linear(d_model, d_model)
        self.w_k = nn.Linear(d_model, d_model)
        self.w_v = nn.Linear(d_model, d_model)

        #  d_model
        self.w_o = nn.Linear(d_model, d_model)

    def forward(self, q, k, v, mask=None):
        """
        
        - q, k, v:  (batch_size, seq_len, d_model)
        - mask: padding

        
        - output:  (batch_size, seq_len, d_model)
        - attn:  (batch_size, num_heads, seq_len, seq_len)
        """

        batch_size = q.size(0)  # batch

        # Step 1Q, K, V  + 
        # (batch_size, seq_len, d_model)
        # (batch_size, seq_len, num_heads, d_k) 
        #       (batch_size, num_heads, seq_len, d_k)
        q = self.w_q(q).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        k = self.w_k(k).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        v = self.w_v(v).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)

        # Step 2 Scaled Dot-Product Attention
        attn_output, attn = scaled_dot_product_attention(q, k, v, mask)
        # attn_output: (batch_size, num_heads, seq_len, d_k)

        # Step 3
        # (batch_size, seq_len, num_heads, d_k)
        #  -> (batch_size, seq_len, d_model)
        attn_output = attn_output.transpose(1, 2).contiguous().view(
            batch_size, -1, self.num_heads * self.d_k
        )

        # Step 4 d_model
        output = self.w_o(attn_output)

        return output, attn
```

## 4. Feed Forward Network
+ ****`(batch, seq, d_model)` 
+ ****
    -  ReLU  dropout
    - `d_ff`
    - `d_model` 
+ ****`(batch, seq, d_model)` 


```python
class FeedForward(nn.Module):
    def __init__(self, d_model, d_ff=2048, dropout=0.1):
        """
        Position-wise Feed-Forward Network

        
        - d_model: Transformer
        - d_ff:  d_model2048
        - dropout: Dropout 
        """
        super().__init__()
        #  d_model d_ff
        self.linear1 = nn.Linear(d_model, d_ff)
        # Dropout 
        self.dropout = nn.Dropout(dropout)
        #  d_ff  d_model
        self.linear2 = nn.Linear(d_ff, d_model)
        #  ReLU
        self.relu = nn.ReLU()

    def forward(self, x):
        """
        
        - x:  (batch_size, seq_len, d_model)

        
        -  (batch_size, seq_len, d_model)
        """
        #   ReLU  Dropout  
        return self.linear2(self.dropout(self.relu(self.linear1(x))))
```

## 5. Encoder Layer
+ ****`(batch, seq, d_model` 
+ ****
    - 
    -  LayerNorm
    - 
    -  LayerNorm
+ ****`(batch, seq, d_model)` 



```python
class EncoderLayer(nn.Module):
    def __init__(self, d_model, num_heads, d_ff=2048, dropout=0.1):
        """
        Transformer Encoder Layer

        
        - d_model: /
        - num_heads: 
        - d_ff: 
        - dropout: dropout
        """
        super().__init__()
        # 
        self.self_attn = MultiHeadAttention(d_model, num_heads)
        # 
        self.ffn = FeedForward(d_model, d_ff, dropout)

        #  LayerNorm 
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)

        #  Dropout 
        self.dropout1 = nn.Dropout(dropout)
        self.dropout2 = nn.Dropout(dropout)

    def forward(self, x, mask=None):
        """
        
        - x:  (batch_size, seq_len, d_model)
        - mask: 

        
        -  (batch_size, seq_len, d_model)
        """
        # =====  =====
        # Q=K=V=x
        attn_out, _ = self.self_attn(x, x, x, mask)
        #  + dropout + LayerNorm
        x = self.norm1(x + self.dropout1(attn_out))

        # =====  =====
        ffn_out = self.ffn(x)
        #  + dropout + LayerNorm
        x = self.norm2(x + self.dropout2(ffn_out))

        return x
```

## 6. Encoder
+ ****`(batch, src_seq, d_model)`
+ ****
    - 
    - 
    - dropout
    - 
+ ****`(batch, src_seq, d_model)`



```python
class Encoder(nn.Module):
    def __init__(self, vocab_size, d_model, num_layers, num_heads, d_ff=2048, max_len=5000, dropout=0.1):
        """
        Transformer 

        
        - vocab_size: 
        - d_model: 
        - num_layers: 
        - num_heads: 
        - d_ff: 
        - max_len: 
        - dropout: dropout
        """
        super().__init__()
        # IDd_model
        self.embedding = nn.Embedding(vocab_size, d_model)
        # 
        self.pos_encoding = PositionalEncoding(d_model, max_len)
        # 
        self.layers = nn.ModuleList([
            EncoderLayer(d_model, num_heads, d_ff, dropout) for _ in range(num_layers)
        ])
        # dropout
        self.dropout = nn.Dropout(dropout)

    def forward(self, src, mask=None):
        """
        
        - src:  (batch_size, seq_len)
        - mask: 

        
        -  (batch_size, seq_len, d_model)
        """
        #  sqrt(d_model) 
        x = self.embedding(src) * math.sqrt(self.embedding.embedding_dim)
        # 
        x = self.pos_encoding(x)
        # dropout
        x = self.dropout(x)
        # 
        for layer in self.layers:
            x = layer(x, mask)
        return x
```

## 7. Decoder Layer
+ ****
+ ****
    - Masked 
    -  LayerNorm
    - -
    -  LayerNorm
    -  LayerNorm
+ ****`(batch, tgt_seq, d_model)`



```python
class DecoderLayer(nn.Module):
    def __init__(self, d_model, num_heads, d_ff=2048, dropout=0.1):
        """
        Transformer Decoder Layer

        
        - d_model: /
        - num_heads: 
        - d_ff: 
        - dropout: dropout
        """
        super().__init__()
        # Masked 
        self.self_attn = MultiHeadAttention(d_model, num_heads)
        # -
        self.cross_attn = MultiHeadAttention(d_model, num_heads)
        # 
        self.ffn = FeedForward(d_model, d_ff, dropout)

        #  LayerNorm 
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        self.norm3 = nn.LayerNorm(d_model)

        #  dropout 
        self.dropout1 = nn.Dropout(dropout)
        self.dropout2 = nn.Dropout(dropout)
        self.dropout3 = nn.Dropout(dropout)

    def forward(self, x, enc_out, src_mask=None, tgt_mask=None):
        """
        
        - x:  (batch_size, tgt_seq_len, d_model)
        - enc_out:  (batch_size, src_seq_len, d_model)
        - src_mask: 
        - tgt_mask: 

        
        -  (batch_size, tgt_seq_len, d_model)
        """
        # ===== masked  =====
        # 
        attn1, _ = self.self_attn(x, x, x, tgt_mask)
        x = self.norm1(x + self.dropout1(attn1))  #  + LayerNorm

        # ===== - =====
        #  query key  value
        attn2, _ = self.cross_attn(x, enc_out, enc_out, src_mask)
        x = self.norm2(x + self.dropout2(attn2))  #  + LayerNorm

        # =====  =====
        ffn_out = self.ffn(x)
        x = self.norm3(x + self.dropout3(ffn_out))  #  + LayerNorm

        return x
```

## 8. Decoder
+ ****
+ ****
    - dropout
    - 
+ ****`(batch, tgt_seq, d_model)`



```python
class Decoder(nn.Module):
    def __init__(self, vocab_size, d_model, num_layers, num_heads, d_ff=2048, max_len=5000, dropout=0.1):
        """
        Transformer 

        
        - vocab_size: 
        - d_model: 
        - num_layers: 
        - num_heads: 
        - d_ff: 
        - max_len: 
        - dropout: dropout
        """
        super().__init__()
        # ID d_model 
        self.embedding = nn.Embedding(vocab_size, d_model)
        # 
        self.pos_encoding = PositionalEncoding(d_model, max_len)
        # 
        self.layers = nn.ModuleList([
            DecoderLayer(d_model, num_heads, d_ff, dropout) for _ in range(num_layers)
        ])
        # dropout 
        self.dropout = nn.Dropout(dropout)

    def forward(self, tgt, enc_out, src_mask=None, tgt_mask=None):
        """
        
        - tgt:  (batch_size, tgt_seq_len)
        - enc_out:  (batch_size, src_seq_len, d_model)
        - src_mask: 
        - tgt_mask: 

        
        -  (batch_size, tgt_seq_len, d_model)
        """
        #  sqrt(d_model) 
        x = self.embedding(tgt) * math.sqrt(self.embedding.embedding_dim)
        # 
        x = self.pos_encoding(x)
        #  dropout
        x = self.dropout(x)
        # 
        for layer in self.layers:
            x = layer(x, enc_out, src_mask, tgt_mask)
        return x
```

## 9. Transformer 
+ ****`src``tgt`
+ ****
    - padding mask
    - padding + mask
    - 
    - 
    - softmax
+ **** logits`(batch, tgt_seq, tgt_vocab_size)`



```python
class Transformer(nn.Module):
    def __init__(self, src_vocab_size, tgt_vocab_size, d_model=512, num_layers=6, num_heads=8, d_ff=2048, max_len=5000, dropout=0.1):
        """
        Transformer

        
        - src_vocab_size: 
        - tgt_vocab_size: 
        - d_model: 
        - num_layers: 
        - num_heads: 
        - d_ff: 
        - max_len: 
        - dropout: dropout
        """
        super().__init__()
        # 
        self.encoder = Encoder(src_vocab_size, d_model, num_layers, num_heads, d_ff, max_len, dropout)
        # 
        self.decoder = Decoder(tgt_vocab_size, d_model, num_layers, num_heads, d_ff, max_len, dropout)
        # softmax
        self.linear_out = nn.Linear(d_model, tgt_vocab_size)

    def make_src_mask(self, src):
        """
        padding mask

        
        - src: (batch_size, src_len) ID

        
        - mask: (batch_size, 1, 1, src_len) bool
          TrueFalsepaddingmask
        """
        mask = (src != 0).unsqueeze(1).unsqueeze(2)  # 
        return mask

    def make_tgt_mask(self, tgt):
        """
        padding maskmask

        
        - tgt: (batch_size, tgt_len) ID

        
        - tgt_mask: (batch_size, 1, tgt_len, tgt_len) bool
          TrueFalsepadding
        """
        batch_size, tgt_len = tgt.size()
        # padding maskTrue
        tgt_mask = (tgt != 0).unsqueeze(1).unsqueeze(2)
        # mask
        nopeak_mask = torch.tril(torch.ones((tgt_len, tgt_len), device=tgt.device)).bool()
        # paddingmask
        tgt_mask = tgt_mask & nopeak_mask
        return tgt_mask

    def forward(self, src, tgt):
        """
        Transformer

        
        - src: (batch_size, src_len) ID
        - tgt: (batch_size, tgt_len) ID

        
        - output: (batch_size, tgt_len, tgt_vocab_size) logits
        """
        # 
        src_mask = self.make_src_mask(src)
        tgt_mask = self.make_tgt_mask(tgt)

        # 
        enc_out = self.encoder(src, src_mask)
        # 
        dec_out = self.decoder(tgt, enc_out, src_mask, tgt_mask)
        # softmax
        output = self.linear_out(dec_out)
        return output
```

```python
#  src_vocab_size = 1000

#  tgt_vocab_size = 1000

#  src_seq_len = 10tgt_seq_len = 9

# batch size = 2


import torch

# 
src_vocab_size = 1000
tgt_vocab_size = 1000
d_model = 512
num_layers = 2
num_heads = 8
batch_size = 2
src_seq_len = 10
tgt_seq_len = 9

# 
model = Transformer(src_vocab_size, tgt_vocab_size, d_model, num_layers, num_heads)

# token[1, vocab_size)
# 0padding token0
src_input = torch.randint(1, src_vocab_size, (batch_size, src_seq_len))
tgt_input = torch.randint(1, tgt_vocab_size, (batch_size, tgt_seq_len))

# 
output = model(src_input, tgt_input)

print(":", output.shape)
#  (batch_size, tgt_seq_len, tgt_vocab_size)
```
